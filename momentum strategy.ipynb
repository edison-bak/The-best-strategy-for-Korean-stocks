```python
'''momentum strategy
buy : 20 stocks that have risen the most in kospi + kosdaq
'''
from pykrx import stock
import FinanceDataReader as fdr
```

```python
'''select 20 stocks that have good momentum(2019.11.01 to 2020.09.29)'''
df1=stock.get_market_ohlcv("20191101")
df2=stock.get_market_ohlcv("20200929")
'''
join() method in pandas is used to merge two DataFrames 
based on their index or a specified column(s)

 The lsuffix and rsuffix parameters are used 
 to add a suffix to the columns of both DataFrames 
 to differentiate them when they have overlapping column names.'''
kospi=df1.join(df2,lsuffix="_l",rsuffix='_r')
kospi
```

```python
kospi=kospi.rename(columns={'종가_r':'(simulation)selling price','종가_l':'(simulation)purchase price'})
kospi['momentum']=100*(kospi['(simulation)selling price']-kospi['(simulation)purchase price'])/kospi['(simulation)purchase price']
kospi=kospi[['(simulation)purchase price','(simulation)selling price','momentum']]
kospi_momentum20=kospi.sort_values(by='momentum',ascending=False)[:20]
kospi_momentum20
```

```python
#check the returns for the selected 20 stocks
df3=stock.get_market_ohlcv("20201102")
df4=stock.get_market_ohlcv("20210430")
pct_df=df3.join(df4,lsuffix="_l",rsuffix="_r")
pct_df
```

```python
pct_df=pct_df[['종가_l','종가_r']]
pct_df.columns=['purchase price','selling price']
kospi_momentum20_result=kospi_momentum20.join(pct_df)
kospi_momentum20_result
```

```python
kospi_momentum20_result['Return']=(kospi_momentum20_result['selling price']/kospi_momentum20_result['purchase price'])
'''
fillna(0) is a method commonly used in pandas, 
a popular data manipulation library in Python. It is used 
to fill missing or NaN (Not a Number) values in a DataFrame 
or Series with a specified value, in this case, with the value 0.'''
average_return=kospi_momentum20_result['Return'].fillna(0).mean()
mom20_cagr=average_return**(1/0.5)-1
mom20_cagr*100
```

```python
# 1 year -> 11year
import pandas as pd
import datetime
from dateutil.relativedelta import relativedelta

year=2010
month=11
period=6

inv_start=f"{year}-{month}-01"

'''two datetime have different role. 
datetime(module).datetime(class).strptime~
if you want to type datetime once.
import datetime-> from datetime import datetime

datetime.strptime(inv_start, "%Y-%m-%d") is a Python code 
that uses the strptime() function from the datetime module 
to parse a date string into a datetime object.'''
inv_start=datetime.datetime.strptime(inv_start,"%Y-%m-%d")
inv_end=inv_start+relativedelta(months=period-1)

mom_start=inv_start-relativedelta(months=12)
mom_end=inv_start-relativedelta(months=2)
print(mom_start.strftime("%Y-%m"),mom_end.strftime("%Y-%m"),"=>",
     inv_start.strftime("%Y-%m"),inv_end.strftime("%Y-%m"))
```

```python
'''the first trading day(index=0) of the input year,month  '''
def get_business_day(df,year,month,index=0):
    str_month=f"{year}-{month}"
    return df.loc[str_month].index[index]
```

```python
df=fdr.DataReader(symbol='KS11')
get_business_day(df,2010,1,0)
```

```python
def momentum(df,year=2010,month=11,period=12):
    #the Investment start date, end date
    str_day=f"{year}-{month}-01"
    start=datetime.datetime.strptime(str_day,"%Y-%m-%d")
    end=start+relativedelta(months=period-1)
    
    # the closing price of the first trading day
    inv_start=get_business_day(df,start.year,start.month,0)
    inv_end=get_business_day(df,end.year,end.month,-1)
    inv_start=inv_start.strftime("%Y%m%d")
    inv_end=inv_end.strftime("%Y%m%d")
    
    #the momentum start date, end date
    end=start-relativedelta(months=2)
    start=start-relativedelta(months=period)
    
    # the closing price of the first momentum trading date 
    mom_start=get_business_day(df,start.year, start.month,0)
    mom_end=get_business_day(df,end.year,end.month,-1)
    mom_start=mom_start.strftime("%Y%m%d")
    mom_end=mom_end.strftime("%Y%m%d")
    print(mom_start,mom_end,"|",inv_start,inv_end)
    df1=stock.get_market_ohlcv_by_ticker(mom_start)
    df2=stock.get_market_ohlcv_by_ticker(mom_end)
    mon_df=df1.join(df2,lsuffix='l',rsuffix="r")
    mon_df['fluctuation rate']=(mon_df['종가r']-mon_df['종가l'])/mon_df['종가l']*100
    
    #Excluding preferred stock
    mon_df=mon_df.filter(regex="0$",axis=0)
    mon20=mon_df.sort_values(by="fluctuation rate",ascending=False)[:20]
    mon20=mon20[['fluctuation rate']]
    
    df3=stock.get_market_ohlcv(inv_start)
    df4=stock.get_market_ohlcv(inv_end)
    inv_df=df3.join(df4,lsuffix="l",rsuffix="r")
    
    inv_df['Return']=inv_df['종가r']/inv_df['종가l']
    inv_df=inv_df[['Return']]
    
    result_df=mon20.join(inv_df)
    result=result_df['Return'].fillna(0).mean()
    return year,result
```
