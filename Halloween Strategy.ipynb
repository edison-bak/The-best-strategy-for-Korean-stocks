```python
def sixmonth_investment(df, start_year=2004, end_year=2023,month=11):
    cumulative_return=1
    
    for year in range(start_year, end_year):
        start=datetime.datetime(year=year, month=month,day=1)
        end=start+relativedelta(months=5)
        
        buy_mon=start.strftime("%Y-%m")
        sell_mon=end.strftime("%Y-%m")
        
        #purchase_price
        purchase_price=df.loc[buy_mon].iloc[0]['Open']
        
        #selling_price
        selling_price=df.loc[sell_mon].iloc[-1]['Close']
        
        #Return
        Return=selling_price/purchase_price
        cumulative_return=cumulative_return*Return
    return cumulative_return
```

```python
sixmonth_investment(kospi,start_year=2004,end_year=2023,month=11)
```

```python
#variations of the buying month(1~12)
for month in range(1,12+1):
    ret=sixmonth_investment(kospi, start_year=2004,end_year=2023,month=month)
# {month:02} is a string formatting expression that formats the variable 
# "month" as a two-digit number with leading zeros. 
    print(f"{month:02} {ret:.2f}")
```

```python
# out put : when i bought the stock in November, the return was the Best.
```

```python
# visualization of the out put

data={}

for month in range(1,12+1):
    ret=sixmonth_investment(kospi,start_year=2004,end_year=2023,month=month)
    data[month]=ret
print(data)
```
